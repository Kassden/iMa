```python
log_estimator = LogisticRegression(multi_class='multinomial',max_iter=10000, solver ='lbfgs',penalty='l2', tol=1e-7)
for_estimator = RandomForestClassifier()

smf_num = SelectFromModel(estimator = log_estimator, max_features=12)
smf_cat = SelectFromModel(estimator = log_estimator, max_features=2)



num_pipeline = Pipeline([
    
    ("imputer", SimpleImputer(strategy = 'mean')),
    ("std_scaler", StandardScaler()),
    ("select_2",SelectKBest(k=26)),
    ("select_3",SelectFpr(alpha=0.05)),
    ("select", smf_num),
    ('pca', PCA(n_components=2)),
    ('poly', PolynomialFeatures(degree=6)),
       
])  

cat_pipeline = Pipeline([
    
    ("imputer", SimpleImputer(strategy = 'most_frequent')),
    ("cat_encoder", OneHotEncoder(handle_unknown = 'ignore')),
    ("select_2",SelectKBest(k=4)),
    ("select_3",SelectFpr(alpha=0.05)),
    ("select", smf_cat),
    
])  


market_pipeline = Pipeline([
    
    ("imputer", SimpleImputer(strategy = 'mean')),
    ("std_scaler", StandardScaler()),
    #('poly', PolynomialFeatures(degree=6)),
    

])  


exo_pipeline = Pipeline([
    
    ("imputer", SimpleImputer(strategy = 'mean')),
    ("std_scaler", StandardScaler()),
    ('pca', PCA(n_components=2))


])
```
