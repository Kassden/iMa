```scala
import spark.implicits._ 
```

```scala
val creditcard = spark.read.option("header",true)
    .option("inferSchema.option("header",true)",true)
    .csv("/Users/mk2/Downloads/creditcard.csv")
```

```scala
creditcard.printSchema()
```

```scala
val Array(train, test) = creditcard.randomSplit(Array(.8, .2), seed=42)
```

```scala
import org.apache.spark.ml.feature.VectorAssembler 

val vecAssembler = new VectorAssembler()
    .setInputCols(Array("Time","V1","V2","V3","V4","V5","V6","V7","V8","V9","V10","V11","V12","V13","V14","V15","V16",
                       "V17","V18","V19","V20","V21","V22","V23","V24","V25","V26","V27","V28","Amount"))
    .setOutputCol("features")

val vec_train = vecAssembler.transform(train)
val vec_test = vecAssembler.transform(test)
 
```

```scala
import org.apache.spark.ml.classification.LogisticRegression

val log_reg = new LogisticRegression()
    .setFeaturesCol("features")
    .setLabelCol("Class")
```

```scala
val log_reg_trained = log_reg.fit(vec_train)
```

```scala
val pred = log_reg_trained.transform(vec_test)
```
