```python
import pandas as pd
import numpy as np

pd.set_option('display.max_column',None)
pd.set_option('display.max_row',500)

import os, json, csv, glob
```

```python
path_to_json = '/Users/mk2/Documents - Gabriel’s MacBook Pro/dadaset' 

json_pattern = os.path.join(path_to_json,'*.json')

file_list = glob.glob(json_pattern)


dfr = []

for file in file_list:
    
    with open(file) as f:  
        
        json_records= pd.json_normalize(json.loads(f.read()),record_path=['records'], meta=['id','name','details'],errors='ignore')
        
    with open(file) as g:
        
        json_mov = pd.json_normalize(json.loads(g.read()),record_path=['movement'], meta=['id','name'])
        
    with open(file) as p:

        json_vet = pd.json_normalize(json.loads(p.read()),record_path=['veterinary'], meta=['id','name'])
    
    with open(file) as z:

        json_track = pd.json_normalize(json.loads(z.read()),record_path=['trackwork'], meta=['id','name'])
        
    dfr.append(json_records), dfm.append(json_mov), dfv.append(json_vet),dft.append(json_track)
    
records, movements, veterinary, trackworks = pd.concat(dfr, sort=False), pd.concat(dfm, sort=False), pd.concat(dfv, sort=False), pd.concat(dft, sort=False)  
    
records_raw = pd.concat(dfr, sort=False)

details = pd.json_normalize(records_raw['details'], meta = ["id","name"])

records_raw.reset_index(drop=True, inplace=True)

records = pd.concat([records_raw, details], axis = 'columns',ignore_index=False)

records.to_csv("/Users/mk2/Documents - Gabriel’s MacBook Pro/dadaset/records.csv")
movemements.to_csv("/Users/mk2/Documents - Gabriel’s MacBook Pro/dadaset/movements.csv")
veterinary.to_csv("/Users/mk2/Documents - Gabriel’s MacBook Pro/dadaset/veterinary.csv")
trackworks.to_csv("/Users/mk2/Documents - Gabriel’s MacBook Pro/dadaset/trackworks.csv")



```
