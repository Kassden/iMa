```python
records["prev_declared_weight"] = records.groupby(['name'])['declared_weight'].shift(periods=-1)

records["prev_actual_weight"] = records.groupby('name')['actual_weight'].shift(periods=-1)

records.loc[:,('prev_speed')] = records.groupby('name')[('speed')].shift(periods=-1)

records.loc[:,('prev_time')] = records.groupby('name')[('finish_time')].shift(periods=-1)

records.loc[:,('prev_result')] = records.groupby('name')[('result')].shift(periods=-1)
```

#Changes

```python
records["declared_wt_change"] = records['declared_weight'] - records['prev_declared_weight'] 

records["actual_wt_change"] = records['actual_weight'] - records['prev_actual_weight'] 

records["prev_speed_change"] = records['prev_speed'] - records['prev_speed'].shift(periods=-1)

```

```python
trackworks["date"] = pd.to_datetime(trackworks["date"],format='%d/%m/%Y')

trackworks["details.age"] = pd.to_numeric(trackworks["details.age"])

trackworks_sorted['name'] = trackworks_sorted['name'].str.strip()

trackworks.dtypes
```

```python
trackworks_sorted = trackworks.sort_values('date',ascending = False)

trackworks_sorted.reset_index(inplace= True)
```

```python
trackworks_sorted.loc[:,('rest_time')] = (trackworks_sorted.groupby('id')["date"].diff(periods=-1).dt.days.fillna(0))
trackworks_sorted.loc[:,('cumul_rest_time')] = trackworks_sorted.groupby('id')['rest_time'].cumsum().fillna(0)
trackworks_sorted.loc[:,('real_age')] = (trackworks_sorted['details.age'] - trackworks_sorted['cumul_rest_time']/365).round()
```
